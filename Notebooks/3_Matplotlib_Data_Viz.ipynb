# Matplotlib Cheat Sheet and Review #

```python
import matplotlib.pyplot as plt
```

```python
#To see Matplotlib in Jupyter
%matplotlib inline
```

```python
#first example
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
# Functional way of plotting
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title("Title")
```

Multiplots on the same canvas

```python
plt.subplot(1,2,1)  #number of rows, columns and plot number
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented Method
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8]) #left of the axis, bottom, width and height of the axes. Must be 0<x<1
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title("Title")
```

```python
#multiplot

fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("Larger Plot")

axes2.plot(y,x)
axes2.set_title("Smaller Plot")
```

***Subplots***

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
# plt.tight_layout()

axes[0].plot(x,y)
axes[0].set_title("First Plot")
axes[1].plot(y,x)
axes[1].set_title("Second Plot")

plt.tight_layout()
```

## Legends ##

```python
fig=plt.figure(dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(y,x**2,label='x-squared')
ax.plot(y,x**3, label='x-cubed')
ax.set_title("Title")
ax.set_ylabel("Y")
ax.set_xlabel("X")

ax.legend(loc=0) #to prevent legend overlapping the plot. See documenation for loc code. 0 is 'best'
# can also specify loc via a tuple. i.e. loc=(0.1,0.1)
```

## Plot Appearance ##

```python
fig = plt.figure(dpi=150)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#9b094b',linewidth=3, alpha=1, linestyle=':') #use either rbg or basic colours
#alpha for transparency
```

```python
fig = plt.figure(dpi=150)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#9b094b',linewidth=3, alpha=1, linestyle='-',marker='o',markersize=10,
       markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='blue') #use either rbg or basic colours
#alpha for transparency

```

## Plot Range and Control Over Axis ##

```python
fig = plt.figure(dpi=150)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#9b094b',linewidth=2,  linestyle=':') #use either rbg or basic colours

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
